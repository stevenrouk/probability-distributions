```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()
```

# Binomial recap
A *Bernoulli trial* is a trial where there are two possible outcomes, where a "success" has probability p and a "failure" has probability (1 - p).

A *Binomial distribution* is when you run multiple Bernoulli trials and count the number of successes.

The `np.random.binomial(n, p, size)` function allows us to sample from a Binomial distribution. The `size` parameter allows us to sample multiple times.

```python
samples = np.random.binomial(100, 0.1, size=10000)
```

```python
_ = plt.hist(samples, bins=22, density=True)
```

What's the probability of getting between 15 and 20 successes out of 100 trials, where the probability of success of each trial is 0.1?

```python
(sum(samples>=15) - sum(samples>20))/len(samples)
```

Let's graph each value on it's own now, rather than binning them in a histogram.

```python
samples_df = pd.DataFrame(samples, columns=["samples"])
```

```python
counts = samples_df["samples"].value_counts()
counts = counts.sort_index()
```

```python
_ = counts.plot(kind='bar')
_ = plt.ylabel('value counts')
```

## Function to run Binomial experiment and plot results
Now let's turn what we just did into a repeatable function that samples from the binomial distribution and plots the results.

```python
def plot_binomial(n, p, size, plot_kind='line'):
    samples = np.random.binomial(n, p, size)
    samples_df = pd.DataFrame(samples, columns=["samples"])
    counts = samples_df["samples"].value_counts()
    counts = counts.sort_index()
    _ = counts.plot(kind=plot_kind)
    _ = plt.ylabel('value counts')
    
    return True
```

Let's blot some binomials for varying n, p, and size.

```python
_ = plot_binomial(100, 0.1, 1000)
```

```python
_ = plot_binomial(100, 0.1, 10000)
```

```python
_ = plot_binomial(100, 0.1, 100000)
```

```python
_ = plot_binomial(100, 0.1, 1000000)
```

```python
_ = plot_binomial(100, 0.1, 10)
```

```python
_ = plot_binomial(100, 0.05, 100000)
```

```python
_ = plot_binomial(10, 0.05, 100000)
```

```python
_ = plot_binomial(10, 0.5, 100000)
```

```python
_ = plot_binomial(10000, 0.5, 100000)
```
